```python
import pandas as pd
from bs4 import BeautifulSoup
import requests


print("Import complete")
```

```python
session = requests.Session()
user = input("Username:")
password = input("Password:")

url = 'http://dmsreports.corp.oncor.com/DMSReports/swagger/ui/index'
r = requests.get(url, auth=(str('corp/'+user),password))
print(r.status_code)
soup = BeautifulSoup(r.text, 'html.parser')
print(r.text)
print("Complete")
```

```python
reports = [str(link) for link in soup.find_all('a',href=True)]
products = ['Home','Current Events','Historical Info','Customer Info','Device Info','Performance','Storm Processes Library','PDF\\Spatial Maps',
              'Help\\External Links','Exports']
exclude = ['Home','Area Summary','Outage Map','DMSSupport@oncor.com', 'New Outage Map ','DMSSupport@oncor.com',
          'DMSSupport@oncor.com', '\r\n                            (888) 841-7796\r\n                        ']

x = 0
reps = []
p = "None"
while x < len(reports):
    title = reports[x].split(">")[1].split("<")[0]
    if title in products:
        p = title
    hypertext_reference = reports[x].split('href="')[-1].split('"')[0]
    if hypertext_reference != 'javascript: void(0);':
        if title not in exclude:
            reps.append([p,title,hypertext_reference])
    x += 1
repdf = pd.DataFrame(reps,columns=['Category','Title','Location'])
with pd.ExcelWriter('Reports.xlsx') as writer:
    repdf.to_excel(writer, sheet_name='Reports',index=False)
    
print("Complete")
```
